```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
# "webdriver.Chrome()"
driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://www.daraz.pk/catalog/?q=bread&_keyori=ss&from=input&spm=a2a0e.searchlist.search.go.2e333e03jhn1iA')
df = pd.DataFrame(columns=['Reviews','Product Name','Price'])
total_reviews=[]
title=[]
prices=[]
sleep(2)

#//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div[1]/div

#//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div[2]/div

for z in range(1,6):
    driver.implicitly_wait(2)
    #if(len((driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div[1]/div/div[1]/div[2]/div['+str(z)+']/div/div/div[2]/div[5]/div/span').text))==0):
     #   continue
    
    #//*[@id="root"]/div/div[3]/div[1]/div/div[1]/div[2]/div[1]/div/div/div[2]/div[5]/div/span
    products = driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(z)+']')
    products.click()
    driver.implicitly_wait(8)
    
    
    #existance=driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[2]/div/span').text
    #if(len(existance)==0):
     #   break
            
    #if(driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]').is_enabled()):
    #ben=True
    #else:
     #   break
    driver.execute_script("window.scrollTo(0,2200)") 
    while True:
        driver.execute_script("window.scrollTo(0,800)") 
        driver.implicitly_wait(20)
        sleep(2)
        for i in range(1,5):
            try:
                pric=driver.find_element_by_xpath('//*[@id="module_product_price_1"]/div/div/span')
                tit=driver.find_element_by_xpath('//*[@id="module_product_title_1"]/div/div/span')
                review = driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[1]/div['+str(i)+']/div[3]/div[1]')
                title.append(tit.text)
                prices.append(pric.text)
                total_reviews.append(review.text)
            except:
                break
        try:
            rep=driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/div').text
        except:
            break
        if(driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]').is_enabled()):
            driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]').click()
        else:
            break
        
    driver.back()
    print("Product "+str(z)+" Reviews Pulled...")
    
    
driver.close()
```

```python
for p in range(len(total_reviews)):
    df.at[p,'Price']=prices[p]
    df.at[p,'Product Name']=title[p]
    df.at[p,'Reviews']=total_reviews[p]
    
    
pd.set_option('display.max_rows', df.shape[0]+1)
```

```python
!pip install python-docx
import docx
negativewords=[]
doc = docx.Document('C:/Users/Abubakar/Desktop/negativeword.docx')
for i in doc.paragraphs:
    #print(i.text)
    negativewords.append(i.text)
    
negativewords
```

```python
positivewords=[]
doc2=docx.Document('C:/Users/Abubakar/Desktop/positiveword.docx')
for i in doc2.paragraphs:
    #print(i.text)
    positivewords.append(i.text)
    
positivewords
```

```python
Sentiments=[]
words=[]
for i in range(len(total_reviews)):
    for j in range(len(negativewords)):
        if negativewords[j] in total_reviews[i]:
            words.append(negativewords[j])
            Sentiments.append("Negative Review")
            break
        if positivewords[j] in total_reviews[i]:
            Sentiments.append("Postive Review")
            words.append(positivewords[j])
            break
        
```

```python
for p in range(len(Sentiments)):
    df.at[p,'Sentiments']=Sentiments[p]
    df.at[p,'Sentimental Words']=words[p]
```
